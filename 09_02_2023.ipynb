## Checking the contents in latest launch column

```python
cs.Latest_Launch.unique()
```

## change its dtype into datetime

```python
cs['Latest_Launch']=pd.to_datetime(cs.Latest_Launch)
cs.head()
```

## Extracting year,month,date from the Latest_launch column

```python
cs['year']=pd.DatetimeIndex(cs.Latest_Launch).year
cs['month']=pd.DatetimeIndex(cs.Latest_Launch).month
cs['day']=pd.DatetimeIndex(cs.Latest_Launch).day
cs.head()
```

## Lets convert the dtype of fuel efficeincy into integer

```python
cs.Fuel_efficiency=cs.Fuel_efficiency.astype(int)
cs.head()
```
