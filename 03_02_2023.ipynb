# PANDAS

```python
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
nf=pd.read_csv("netflix_titles_2021.csv")
nf
```

# assign 
###### assigning new columns with certain condition

```python
nf.assign(after_2020=nf['release_year']>2020)
```

```python
#assigning new column method2
nf['after_2019']=nf['release_year']>2019
nf
```

# QUERY
##### TO FIND CERTAIN ROWS AND COLUMNS

```python
nf.query('release_year==2020 & type=="Movie"')
```

```python
nf.isnull().sum()
```

```python
nf=nf.dropna(subset=['director','cast','country','date_added','rating','duration'])
nf
```

```python
nf.drop(columns=['cast','description'])
```

```python
pivot=pd.pivot_table(columns='rating',values='release_year',data=nf,aggfunc='median')
pivot
```
